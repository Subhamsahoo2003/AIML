```python
## Ensure 'Age_mean' exists before using it
df['Age_mean'] = df['Age'].fillna(df['Age'].mean())
df['value_new'] = df['Age_mean'].astype('float')
df.head()
```

```python
df["Value"]=df['value_new'].fillna(df['value_new'].mean()).astype(int)
df
```

```python
df['value']=df['Value'].apply(lambda x:x*2)
df.head()
```
